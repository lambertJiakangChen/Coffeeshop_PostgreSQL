```python
%sql SELECT DISTINCT SHOP_NAME FROM employees;
```

```python
%sql SELECT ROUND(AVG(salary), 0), MIN(salary), MAX(salary) FROM employees;
```

```python
%sql SELECT shop_name, COUNT(*) FROM employees GROUP BY shop_name ORDER BY COUNT(*) DESC;
```

```python
%sql SELECT shop_name, COUNT(*) as shop_numbers, ROUND(AVG(salary), 0) as ave_salary, MIN(salary) as lowest_salary, MAX(salary) as highest_salary FROM employees GROUP BY shop_name ORDER BY COUNT(*) ASC;
```

```python
%sql SELECT e.employee_id, e.shop_name, e.salary FROM employees e WHERE e.shop_name = 'Early Rise' AND e.salary IN (SELECT MAX(salary) FROM employees GROUP BY shop_name);
```

```python
%sql SELECT * FROM employees WHERE email LIKE '%.uk'
```

```python
%sql SELECT first_name, last_name, shop_name, salary, ROUND(average_by_shop, 2) AS average_by_shop from (SELECT first_name, last_name, shop_name, salary, AVG(salary) OVER (PARTITION BY shop_name) AS average_by_shop FROM employees) AS amumu;
```

```python
%sql SELECT first_name, last_name, shop_name, salary, ROUND(average_by_shop, 2) AS average_by_shop, CASE WHEN salary > average_by_shop THEN 'Yes' ELSE 'No' END AS above_average from (SELECT first_name, last_name, shop_name, salary, AVG(salary) OVER (PARTITION BY shop_name) AS average_by_shop FROM employees) AS amumu;
```

```python
%sql SELECT first_name, last_name, shop_name, salary, RANK() OVER (PARTITION BY shop_name ORDER BY salary DESC) as salary_rank, RANK() OVER (ORDER BY salary DESC) AS overall_rank FROM employees ORDER BY shop_name;
```

```python
%sql SELECT first_name, last_name, shop_name, salary, hire_date, RANK() OVER (PARTITION BY shop_name ORDER BY salary DESC) as salary_rank, LAG(salary) OVER(PARTITION BY shop_name) FROM employees
```
